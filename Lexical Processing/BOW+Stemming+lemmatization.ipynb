```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

### BOW on sample three documents

```python
docs = ["This is a great movie.", "The success of a movie depends on the performance of the actors.", "There are no new movies releasing this week."]
print(docs)
```

```python
def preprocess(doc):
  doc = doc.lower()
  words = word_tokenize(doc)
  words = [word for word in words if word not in stopwords.words("english")]
  doc = " ".join(words)
  return doc
```

```python
docs = [preprocess(doc) for doc in docs]
print(docs)
```

```python
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(docs)
print(bow)  # returns cells which have value 1
```

```python
print(bow.toarray())
```

```python
print(bow.shape)
print(vectorizer.get_feature_names_out())
```

### Stemming and lemmatization

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(doc,stem = True):
  doc = doc.lower()
  words = word_tokenize(doc)
  words = [word for word in words if word not in stopwords.words("english")]
  if stem:
    words = [stemmer.stem(word) for word in words]
  else:
    words = [lemmatizer.lemmatize(word,pos = 'v') for word in words]
  
  doc = " ".join(words)
  return doc
```

###BOW on stemmed messages

```python
msgs = [preprocess(msg,stem=True) for msg in spam.message]
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(msgs)
```

```python
pd.DataFrame(bow.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
print(vectorizer.get_feature_names_out())
```

### BOW on lemmatized messages

```python
msgs = [preprocess(msg,stem=False) for msg in spam.message]
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(msgs)
```

```python
pd.DataFrame(bow.toarray(),columns = vectorizer.get_feature_names_out())
```

```python
print(vectorizer.get_feature_names_out())
```
